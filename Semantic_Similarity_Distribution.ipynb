```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 读取CSV文件
df = pd.read_csv('Semantic_Similarity_result.csv')

# 绘制创意评分的分布图
plt.figure(figsize=(10, 6))
sns.histplot(df['Creative Score'], bins=30, kde=True)
plt.title('Distribution of Creative Scores')
plt.xlabel('Creative Score')
plt.ylabel('Frequency')
plt.savefig('creative_score_distribution.png')

# 显示绘图
plt.show()

# 输出描述性统计信息
desc_stats = df['Creative Score'].describe()
print(desc_stats)
```

```python
df_filtered = df[df['Creative Score'] != 1]
```
